```python
features = training_data.drop(['hz_label'], axis = 1)
labels = training_data.hz_label
```

```python
# Search the optimal K for KNN classifier
def search_optimal_k(features,
                     labels,
                     start_k,
                     end_k,
                     stand_scaler,
                     split_and_train_iteration_count,
                     test_data_percentage,
                     shuffle_during_split=True,
                     show_progress=True):
    print(f"Searching the optimal k in between {start_k} ~ {end_k} \n\nIn progress...")

    mean_error_rates_by_ks = []
    for i in range(start_k, end_k + 1):
        mean_error_rates_specific_k = []
        for j in range(split_and_train_iteration_count):
            features_train, features_test, labels_train, labels_test = train_test_split(features,
                                                                                        labels,
                                                                                        test_size=test_data_percentage,
                                                                                        random_state=0,
                                                                                        shuffle=shuffle_during_split)

            features_train_sc = stand_scaler.fit_transform(features_train)
            features_test_sc = stand_scaler.transform(features_test)

            knn_classifier = KNeighborsClassifier(n_neighbors=i,
                                                  metric='minkowski',
                                                  p=2)
            knn_classifier.fit(features_train, labels_train)
            mean_prediction_error_rate = np.mean(knn_classifier.predict(features_test) != labels_test)
            mean_error_rates_specific_k.append(mean_prediction_error_rate)

        mean_error_rates_by_ks.append(np.mean(mean_error_rates_specific_k))
        if show_progress == True:
                print(f'For KNN classifier with K = {i}, mean prediction error rate = {np.mean(mean_error_rates_specific_k):.3}')

    optimal_k, optimal_mean_error_rate = mean_error_rates_by_ks.index(min(mean_error_rates_by_ks))+1, min(mean_error_rates_by_ks)
    print(f'\nDone! The Optimal K for KNN classifier is: {optimal_k}\n')

    # Display error rate v.s. K
    plt.figure(figsize = (7,5))
    plt.plot(range(start_k, end_k + 1),
             mean_error_rates_by_ks,
             'mo--',
             markersize=8,
             markerfacecolor='c',
             linewidth=1)

    # Mark the optimal K for KNN classifier
    plt.plot(optimal_k,
             optimal_mean_error_rate,
             marker='o',
             markersize=8,
             markerfacecolor='gold',
             markeredgecolor='g')

    plt.title(f"The Optimal K for KNN classifier is: {optimal_k}",
              color='blue',
              fontsize=15)
    plt.ylabel("Mean Error Rate",
               color='black',
               fontsize = 13)
    plt.xlabel("K",
               color='black',
               fontsize = 13)

    return optimal_k
```

```python
# Split data with 70% for training and 30% for testing
split_test_data_percentage = 0.3

# Number of rounds to split into training & testing data and train the decision tree for each max depth to search for the optimal max depth
split_and_train_iteration_count = 100

stand_scaler = StandardScaler()

# Search for the optimal mx depth for decision tree
optimal_k = search_optimal_k(features=features,
                             labels=labels,
                             start_k=1,
                             end_k=20,
                             stand_scaler=stand_scaler,
                             split_and_train_iteration_count=split_and_train_iteration_count,
                             test_data_percentage=split_test_data_percentage,
                             shuffle_during_split=True,
                             show_progress=True)
```

```python
# Split data with 70% for training and 30% for testing
features_train, features_test, labels_train, labels_test = train_test_split(features,
                                                                            labels,
                                                                            test_size=split_test_data_percentage,
                                                                            random_state=0)
```

```python
# Standadize the scales of features
sc = StandardScaler()
features_train_sc = sc.fit_transform(features_train)
features_test_sc = sc.transform(features_test)
```

```python
# Train a KNN classifier with the training data
knn_classifier = KNeighborsClassifier(n_neighbors=optimal_k,
                                      metric='minkowski',
                                      p=2)
knn_classifier.fit(features_train_sc, labels_train)
```

```python
# Call trained KNN classifier to get predictions for the testing data
labels_pred = knn_classifier.predict(features_test_sc)
```

```python
# Calculate accuracy, precision, recall, and F-1 scores for the KNN classifier
print("KNN Classifier Accuracy: ", accuracy_score(labels_test, labels_pred))
print()
print("KNN Classifier Classification Report :\n",(classification_report(labels_test,labels_pred)))
```
